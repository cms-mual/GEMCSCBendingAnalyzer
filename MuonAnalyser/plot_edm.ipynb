```python
import ROOT as r
```

```python
from DataFormats.FWLite import Events, Handle
```

```python
file = r.TFile("example_edm.root")
file
```

```python
handle  = Handle('std::vector<reco::Muon>')
label = "muons1Leg"
handle_refit  = Handle('std::vector<reco::Track>')
label_refit = "TrackRefitter"
```

```python
events = Events ("example_edm.root")
for event in events:
    event.getByLabel(label,handle)
    event.getByLabel(label_refit,handle_refit)
    muons = handle.product()
    refit_tracks = handle_refit.product()
    
    print("------new event------")
    for muon in muons:
        print("muon", muon.pt())
    for track in refit_tracks:
        print("refit", track.pt())
```
